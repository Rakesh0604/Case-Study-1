# Case study on Presentation of data

```python
import pandas as pd
data = pd.read_csv('cars_data.csv')
data
```

```python
data.isna().sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.bar(data['Buyer Gender'],height=10000,width=0.8)
plt.xlabel('Buyer Gender')
plt.ylabel('Count')
plt.title('Male VS Female Buyers')
plt.show()
```

```python
top5=data.nlargest(5,['Resell Price'])
print(top5['Make'])
```

```python
least5=data.nsmallest(5,['Resell Price'])
print(least5['Make'])
```
